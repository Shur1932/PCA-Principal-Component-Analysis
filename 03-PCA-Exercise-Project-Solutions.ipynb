Перехожу к методу главных компонент.

## Масштабирование данных

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_pixels = scaler.fit_transform(pixels)
```

```python
scaled_pixels
```

## Метод главных компонент (PCA - Principal Component Analysis)

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

**Выясню, сколько вариативности объясняется этими 2 главными компонентами.**

```python
np.sum(pca_model.explained_variance_ratio_)
```

**Построю график scatterplot для цифр в 2-мерном пространстве главных компонент, раскрашивая график разными цветами на основе колонки number_label в исходном датафрейме.**

```python
plt.figure(figsize=(10,6),dpi=150)
labels = digits['number_label'].values
sns.scatterplot(pca_pixels[:,0],pca_pixels[:,1],hue=labels,palette='Set1')
plt.legend(loc=(1.05,0))
```
